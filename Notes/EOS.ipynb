## To Geometrized

```python
c = 2.99792458e+10   # Speed of light (cm s^{-1})
G = 6.6742e-8        # Gravitational constants (cm^3 g^{-1} s^{-2})
M = 1.98848e+33      # Solar mass (g)
R = 6.9551E+10       # Solar radius (cm)
```

```python
rho_c_geom = 3.48E+15 * R^2 * G / c^2
```

```python
rho_c_geom
```

```python
log(rho_c_geom, 10)
```

```python
log_p_geom = zeta(log(rho_c_geom, 10) + log(c^2 / (G * R^2), 10)) - log(c^4 / (R^2 * G), 10); log_p_geom
```

```python
plot(zeta(xi + log(c^2 / (G * R^2), 10)) - log(c^4 / (R^2 * G), 10), (xi, -0.5, 9.5))
```

```python
diff(zeta(xi + log(c^2/(G*R^2), 10)) - log(c^4/(R^2*G), 10), xi)(log(rho_c_geom, 10))
```

```python
zeta(log(rho_c_geom, 10) + log(c^2/(G*R^2), 10))
```

```python
new_zeta(xi) = zeta(xi + log(c^2 / (G * R^2), 10)) - log(c^4 / (R^2 * G), 10)
```

```python
new_zeta
```

```python
dzeta(xi) = diff(new_zeta(xi), xi)
```

```python
dzeta
```

```python
var('r m rho')
```

```python
p(rho) = 10^(new_zeta(log(rho, 10)))
```

```python
p
```

```python
tov(rho) = -m*rho/r^2 * (1 + p(rho)/rho) * (1 + 4*pi*r^3*p(rho)/m) / (1 - 2*m/r)
```

```python
P = desolve_system_rk4([4*pi*rho*r^2, rho/p(rho) * tov(rho) / dzeta(log(rho, 10)) * 1 / log(new_zeta(log(rho,10)), 10)], [m, rho], ics=[1e-15, 4*pi*1e-15^2*rho_c_geom, rho_c_geom], ivar=r, end_points=100)
```
